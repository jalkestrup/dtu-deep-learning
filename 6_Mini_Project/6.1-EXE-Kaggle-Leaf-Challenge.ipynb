## Exercise 6
6.1) Explain shortly why the size of batches is important. You should comment on how the size of batches affect the memory, training speed, and the estimates of the gradients.
Choose a reasonable batch size the `batch_size = ?` parameter below, and note your choice.

 * **Answer:**
 Batch size determines how many samples are processed per forward pass in parallel. As activations from each layer need to be stored for backpropagation, a larger batch size requires more memory. However, a larger batch size will also speed up the training process since more samples are processed at once. In addition to factors like training speed and hardware limitations, batch size also affects the performance of the model due to its influence on gradient estimates. Generally, a larger batch size results in smoother and more stable gradient updates (since a wider range of samples is averaged when calculating the loss). On the other hand, a smaller batch size can act as an effective form of regularization by introducing more variance into the optimization steps.

 Our dataset is a bit unusual, in the sense that there are 99 classes but only 10 samples per class (very few). Because of the few samples, there is a high risk of overfitting (without considering data augmentation just yet). A large batch size would typically produe a smother and lower variance gradients, which could be problematic if the network learns to memorize the few examples rather than generalize on unseen data. For this reason, I would opt for a rather low batch size, in the range of 8-32.

````python
height, width = IMAGE_SHAPE
channels = 1

conv_out_channels = 6  # <-- Filters in your convolutional layer
kernel_size = 3        # <-- Kernel size
conv_stride =  1      # <-- Stride
conv_pad    =  1      # <-- Padding

#Margin, shape and texture features
mst_input_size = NUM_FEATURES
gru_hidden_size = 128
gru_num_layers = 1
raw_num_features = 2

conv_out_feat = height*width*conv_out_channels
gru_out_feat = gru_hidden_size
raw_out_feat = NUM_FEATURES * raw_num_features
features_cat_size = conv_out_feat+gru_out_feat+raw_out_feat #<-- Number of features concatenated before output layer

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.conv_1 = Conv2d(in_channels=channels, #Greyscale, 1 channel
                             out_channels=conv_out_channels, #Number of filters
                             kernel_size=kernel_size, #
                             stride=conv_stride,
                             padding=conv_pad)

        # Exercise: Add a recurrent unit like and RNN or GRU
        # >> YOUR CODE HERE <<
        self.rnn_1 = GRU(input_size=mst_input_size,
                         hidden_size=gru_hidden_size,
                         num_layers=gru_num_layers,
                         batch_first=True)

        self.l_out = Linear(in_features=features_cat_size,
                            out_features=NUM_CLASSES,
                            bias=False)

    def forward(self, x_img, x_margin, x_shape, x_texture):
        features = []
        out = {}

          ## Convolutional layer ##
        # Convert from (batch_size, height, width, channels) to (batch_size, channels, height, width)
        #x_img = x_img.permute(0, 3, 1, 2)  # Permute to (batch_size, 1, 64, 64)

        # CNN for the images
        x_img = self.conv_1(x_img)          # Apply Conv2d
        x_img = torch.relu(x_img)           # Apply activation function
        features_img = x_img.flatten(start_dim=1)  # Flatten the output for concatenation

        # Append image features to the list
        features.append(features_img)

        # GRU path for the sequential features...

        # Stack and pass through the GRU
        x_seq = torch.stack((x_margin, x_shape, x_texture), dim=1)  # Shape: (batch_size, 3, 64)
        gru_out, _ = self.rnn_1(x_seq)                             # GRU output: (batch_size, 3, hidden_size)
        features_rnn = gru_out[:,-1,:]                             # Select the last hidden state

        # Append GRU features
        features.append(features_rnn)

        ## Use concatenated leaf features for FFNN (e.g., margin and texture vectors) ##
        x = torch.cat((x_margin, x_texture), dim=1)  # Concatenate margin and texture
        features.append(x)                           # Append this concatenation

        ## Output layer where all features are in use ##
        features_final = torch.cat(features, dim=1)   # Concatenate all features
        out['out'] = self.l_out(features_final)       # Fully connected output layer
        return out

net = Net()
if use_cuda:
    net.cuda()
print(net)
````

````python
import torchvision.transforms as T
from PIL import Image

def numpy_to_pil(image):
    """ Converts a numpy array with values [0.0, 1.0] to a PIL Image (uint8). """
    # Scale from [0,1] float32 to [0,255] uint8 for PIL compatibility:
    return Image.fromarray((image * 255).astype(np.uint8).squeeze())

def pil_to_numpy(image):
    """ Converts a PIL image back to a numpy array and re-scales to original [0, 1] range. """
    return np.array(image).astype(np.float32) / 255.0  # Convert back to float32 and scale to range [0, 1]

# Augmentation pipeline
img_transform = T.Compose([
    T.RandomHorizontalFlip(p=0.5),             # Horizontally flip with 50% probability
    T.RandomVerticalFlip(p=0.5),               # Vertically flip with 50% probability
    T.RandomRotation(degrees=2),              # Rotate randomly within a small angle range (-15 to 15 degrees)
    #T.RandomResizedCrop(size=(64, 64), scale=(0.8, 1.0)),  # Random crop and resize
    #T.RandomAffine(degrees=3, shear=2),       # Apply random affine transformation (rotation, shear)
    T.ToTensor(),                              # Convert to PyTorch tensor
])
````

````python
# Setup settings for training
VALIDATION_SIZE = 0.1 # 0.1 is ~ 100 samples for validation
max_iter = 5000
log_every = 100
eval_every = 100

# Early stopping settings
patience = 10   # Number of validations to wait for improvement
best_val_loss = float('inf')  # Initialize to very high value
no_improvement_count = 0      # Counter to keep track of epochs without improvement

############################ Reset network on every re-run
# Initialize the network
net = Net()

if use_cuda:
    net.cuda()  # Move the network to GPU if available

# Initialize a fresh optimizer
learning_rate = 0.001
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate, weight_decay=0.001)

# Initialize the loss function
#criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
criterion = nn.CrossEntropyLoss()

# Clear previous training logs
train_iter = []
train_loss, train_accs = [], []
valid_iter = []
valid_loss, valid_accs = [], []
############################ END

# Function to get label
def get_labels(batch):
    return get_variable(Variable(torch.from_numpy(batch['ts']).long()))

"""
# Function to get input
def get_input(batch):
    return {
        'x_img': get_variable(Variable(torch.from_numpy(batch['images']))),
        'x_margin': get_variable(Variable(torch.from_numpy(batch['margins']))),
        'x_shape': get_variable(Variable(torch.from_numpy(batch['shapes']))),
        'x_texture': get_variable(Variable(torch.from_numpy(batch['textures'])))
    }
"""

def get_input(batch, data_augment=True):
    if data_augment:
        # Apply augmentation transform to each image in the batch for training data only
        images = [pil_to_numpy(img_transform(numpy_to_pil(image))) for image in batch['images']]
    else:
        # Leave validation/test images as they are (numpy arrays, no modification)
        images = batch['images']

    # Now, continue with the same existing conversion to tensor using torch.from_numpy
    return {
        'x_img': get_variable(Variable(torch.from_numpy(np.array(images)))),
        'x_margin': get_variable(Variable(torch.from_numpy(batch['margins']))),
        'x_shape': get_variable(Variable(torch.from_numpy(batch['shapes']))),
        'x_texture': get_variable(Variable(torch.from_numpy(batch['textures'])))
    }


# Generate batches
batch_gen = data_utils.batch_generator(data,
                                       batch_size=batch_size,
                                       num_classes=NUM_CLASSES,
                                       num_iterations=max_iter,
                                       seed=42,
                                       val_size=VALIDATION_SIZE)

# Train network
net.train()
for i, batch_train in enumerate(batch_gen.gen_train()):
    if i % eval_every == 0:

        # Do the validaiton
        net.eval()
        val_losses, val_accs, val_lengths = 0, 0, 0
        for batch_valid, num in batch_gen.gen_valid():
            output = net(**get_input(batch_valid, data_augment=False))
            labels_argmax = torch.max(get_labels(batch_valid), 1)[1]
            val_losses += criterion(output['out'], labels_argmax) * num
            #print(f"Batch {i}: Loss {val_losses.item()}")
            val_accs += accuracy(output['out'], labels_argmax) * num
            val_lengths += num

        # Divide by the total accumulated batch sizes
        val_losses /= val_lengths
        val_accs /= val_lengths
        valid_loss.append(get_numpy(val_losses))
        valid_accs.append(get_numpy(val_accs))
        valid_iter.append(i)

        # Early stopping: Check if the validation loss improved
        if val_losses < best_val_loss:
            best_val_loss = val_losses  # Update best validation loss
            no_improvement_count = 0  # Reset the counter since we have improvement
        else:
            no_improvement_count += 1  # Increment the counter if no improvement

        # If the validation loss hasn't improved for 'patience' evaluations, stop the training
        if no_improvement_count >= patience:
            #print(f"Early stopping at iteration {i}, Best Validation Loss: {best_val_loss}")
            break

        #print("Valid, it: {} loss: {:.2f} accs: {:.2f}\n".format(i, valid_loss[-1], valid_accs[-1]))
        net.train()

    # Train network
    output = net(**get_input(batch_train, data_augment=False))
    labels_argmax = torch.max(get_labels(batch_train), 1)[1]
    batch_loss = criterion(output['out'], labels_argmax)

    train_iter.append(i)
    train_loss.append(float(get_numpy(batch_loss)))
    train_accs.append(float(get_numpy(accuracy(output['out'], labels_argmax))))

    optimizer.zero_grad()
    batch_loss.backward()

    # Added gradient clip as per recommendation from lecture video
    #torch.nn.utils.clip_grad_norm_(net.parameters(), max_norm=1.0)  # Clipping gradients by norm


    optimizer.step()

    # Log i figure
    if i % log_every == 0:
        fig = plt.figure(figsize=(12,4))
        plt.subplot(1, 2, 1)
        plt.plot(train_iter, train_loss, label='train_loss')
        plt.plot(valid_iter, valid_loss, label='valid_loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(train_iter, train_accs, label='train_accs')
        plt.plot(valid_iter, valid_accs, label='valid_accs')
        plt.legend()
        plt.show()
        clear_output(wait=True)
#         print("Train, it: {} loss: {:.2f} accs: {:.2f}".format(i, train_loss[-1], train_accs[-1]))

    if max_iter < i:
        break
````
